```python
x = [15, 18, 20]
y = [22, 26, 16]
x2 = [6, 9, 11]
y2 = [6, 15, 7]
plt.plot(x, y, "r", label = "line 1", linewidth = 5);
plt.plot(x2, y2, "g", label = "line 2", linewidth = 8);
plt.title("Graph")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend();
plt.grid()
```

```python
# bar graph
plt.bar([0.25, 1.25, 2.25, 3.25, 4.25], [50, 40, 70, 80, 20],
       label = "BMW", color = "r", width = 0.1)
plt.bar([0.75, 1.75, 2.75, 3.75, 4.75], [80, 20, 20, 50, 60],
       label = "Audi", color = "b", width = 0.5)
plt.legend()
plt.xlabel("Days")
plt.ylabel("Distance (km)")
plt.title("Cars")
plt.show()
```

```python
import pandas as pd
iris = pd.read_csv(r"C:\Users\hp\Downloads\Iris.csv")
iris.head()
```

```python
for species, irissubset in iris.groupby("Species"):
    plt.scatter(irissubset["PetalLengthCm"], irissubset["PetalWidthCm"], label = species)
plt.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")
for species, irissubset in iris.groupby("Species"):
    ax.scatter(irissubset["PetalLengthCm"], irissubset["PetalWidthCm"], irissubset["SepalWidthCm"], label = species)
plt.legend()
ax.set_xlabel("PetalLength")
ax.set_ylabel("PetalWidth")
ax.set_zlabel("SepalWidth")
plt.show()
```

```python
slices = [2, 2, 2, 13]
activities = ["sleeping", "eating", "working", "playing"]
cols = ["c", "m", "r", "b"]
plt.pie(slices, labels = activities, colors = cols, startangle = 180)
plt.title("Pie plot")
plt.show()
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)
plt.subplot(211) # no of plots, row, column
plt.plot(t1, f(t1), "bo", t2, f(t2))
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2))
plt.show()
```
